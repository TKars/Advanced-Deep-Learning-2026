### <font color='DarkOrange'> **Разработать загрузчик данных [3 балла]**</font>

````python
class RecognitionDataset(Dataset):
    # Базовый алфавит (A-Z, 0-9) начинается с индекса 3
    _default_base_alphabet = { item : index + 3 for index, item in enumerate( digits + ascii_uppercase ) }

    # Специальные токены: начало, конец последовательности и паддинг
    _default_util_alphabet = {'<SOS>': 0, '<EOS>': 1, '<PAD>': 2 }

    # Объединение служебных и базовых токенов в полный алфавит
    _default_full_alphabet = { **_default_util_alphabet, **_default_base_alphabet }

    # Преобразования изображений по умолчанию
    _default_transform = Compose([
        ToImage(),  # Преобразование в формат PIL to tensor
        ToDtype(torch.float32, scale=True),  # Приведение к float32 и масштабирование в [0, 1]
        Resize(size=(32, 160)),  # Масштабирование изображения к заданному размеру
        Normalize(mean=[0.485, 0.456, 0.406],  # Нормализация (значения, как в ImageNet)
                  std=[0.229, 0.224, 0.225])
    ])
    @staticmethod
    def _compute_mask(text):
        """
        Возвращает маску по тексту:
        - 'D' для цифры
        - 'L' для заглавной буквы
        Пример: 'A123BC45' → 'LDDDLLDD'
        Если в тексте есть недопустимые символы, возвращает None
        """
        mask = []
        for char in text:
            if char in digits:
                mask.append("D")
            elif char in ascii_uppercase:
                mask.append("L")
            else:
                return None
        return "".join(mask)

    @staticmethod
    def _filter_markup(source_markup, alphabet, mask_s):
        """
        Фильтрует исходную разметку:
        - Только символы, входящие в алфавит
        - Только строки, соответствующие допустимым шаблонам маски
        """
        target_markup = []
        for item in tqdm(source_markup):
            text = item["text"]
            mask = RecognitionDataset._compute_mask(text)
            if np.all([char in alphabet for char in text]) and (mask in mask_s):
                target_markup.append({"file": item["file"], "text": item["text"]})
        return target_markup

    def _process_markup(self, markup_filepath, images_path, alphabet, mask_s):
        """
        Загружает и фильтрует аннотации, добавляет путь к файлам изображений
        """
        files = os.listdir(images_path)
        texts = [filename.split('_')[-1].split('.')[0] for filename in files]

        source_markup = [{'file': file, 'text': text} for file, text in zip(files, texts)]
        filtered_markup = RecognitionDataset._filter_markup(source_markup, alphabet, mask_s)

        target_markup = []
        for item in filtered_markup:
            target_markup.append({
                "file": os.path.join(images_path, item["file"]),  # Полный путь к изображению
                "text": item["text"]
            })

        return target_markup

    def text_to_tensor(self, text):
        """
        Преобразует строку в тензор индексов:
        [<SOS>, char1, char2, ..., charN, <EOS>]
        Недопустимые символы заменяются на <PAD>
        Обрезается до max_seq_len
        """
        index_s = [self.alphabet['<SOS>']] + \
                  [self.alphabet.get(ch, self.alphabet['<PAD>']) for ch in text.upper()] + \
                  [self.alphabet['<EOS>']]
        return torch.tensor(index_s[:self.max_seq_len], dtype=torch.long)

    def __init__(self, images_path, markup_filepath, transform=None, alphabet=None, max_seq_len=11):
        """
        Инициализация датасета:
        - images_path:     путь к изображениям
        - markup_filepath: путь к JSON-разметке
        - transform:       преобразования изображений
        - alphabet:        словарь символ → индекс
        - max_seq_len:     максимальная длина последовательности
        """
        self.markup_filepath = markup_filepath
        self.images_path     = images_path
        self.max_seq_len     = max_seq_len

        self.transform = transform or self._default_transform
        self.alphabet  = alphabet  or self._default_full_alphabet

        # Допустимые шаблоны для текста (например, маска для госномера)
        self.mask_s = ["LDDDLLDD", "LDDDLLDDD"]

        # Предобработка аннотаций и сохранение валидных примеров
        self.sample_s = self._process_markup(markup_filepath, images_path, self.alphabet, self.mask_s)

    def __len__(self):
        """
        Возвращает:
        - Общий размер набора данных
        """
        return len(self.sample_s)

    def __getitem__(self, idx):
        """
        Возвращает:
        - Преобразованное изображение
        - Тензор индексов текста
        - Исходный текст (для удобства)
        """
        # TODO

        # converted img
        sample = self.sample_s[idx]
        
        img = Image.open(sample['file']).convert('RGB')
        target_image = self._default_transform(img)

        # text -> indices
        text = sample['text']
        label_tensor = self.text_to_tensor(text)

        return target_image, label_tensor, text


class recognition_collate_fn:
    def __init__(self, pad_index):
        self.pad_index = pad_index
    def __call__(self, batch):
        image_s, label_s, text_s = zip(*batch)

        image_s = torch.stack(image_s)

        ### TODO
        padded_labels = pad_sequence(label_s, padding_value=self.pad_index).T
        
        return image_s, padded_labels, text_s
````

#### Энкодер модели

**Реализовать класс `Encoder` согласно архитектуре. Энкодер будет состоять из сверточных, рекуррентного и полносвязного слоев. Отметим что мы решаем задачу `seq2seq` и в данном случае не важно, как мы обрабатываем последовательность слева-направо или наборот поэтому в слое `LSTM` `bidirectional`==True**

````python
from torch.nn import Module, GRU, LSTM, Sequential, Conv2d, Linear
class Encoder(nn.Module):
    def __init__(self, imgH, nH):
        super(Encoder, self).__init__()
        assert imgH % 16 == 0, 'imgH has to be a multiple of 16'

        self.cnn = nn.Sequential(
            nn.Conv2d(3, 64, 3, 1, 1), nn.PReLU(),
            nn.MaxPool2d(2, 2),

            nn.Conv2d(64, 128, 3, 1, 1), nn.PReLU(),
            nn.MaxPool2d(2, 2),

            nn.Conv2d(128, 256, 3, 1, 1), nn.PReLU(),
            nn.BatchNorm2d(256), nn.PReLU(),

            nn.Conv2d(256, 256, 3, 1, 1), nn.PReLU(),
            nn.MaxPool2d(2, 2),

            nn.Conv2d(256, 512, 3, 1, 1),
            nn.BatchNorm2d(512), nn.PReLU(),

            nn.Conv2d(512, 512, 3, 1, 1), nn.PReLU(),
            nn.MaxPool2d((2, 1), (2, 1)),

            nn.Conv2d(512, 512, (2, 1), 1, 0),
            nn.BatchNorm2d(512), nn.PReLU()
        )

        self.rnn = nn.LSTM(512, nH, bidirectional=True)
        self.linear = nn.Linear(2 * nH, nH)

    def forward(self, input):
        conv = self.cnn(input)
        conv = conv.squeeze(2)
        conv = conv.permute(2, 0, 1)

        recurrent, _ = self.rnn(conv)
        output = self.linear(recurrent)

        return output
````

````python
class Encoder(nn.Module):
    def __init__(self, imgH, nH):
        super(Encoder, self).__init__()
        assert imgH % 16 == 0, 'imgH has to be a multiple of 16'

        self.cnn = nn.Sequential(
            nn.Conv2d(3, 64, 3, 1, 1), nn.PReLU(),
            nn.MaxPool2d(2, 2),

            nn.Conv2d(64, 128, 3, 1, 1), nn.PReLU(),
            nn.MaxPool2d(2, 2),

            nn.Conv2d(128, 256, 3, 1, 1),
            nn.BatchNorm2d(256), nn.PReLU(),

            nn.Conv2d(256, 256, 3, 1, 1), nn.PReLU(),
            nn.MaxPool2d(2, 2),

            nn.Conv2d(256, 512, 3, 1, 1),
            nn.BatchNorm2d(512), nn.PReLU(),

            nn.Conv2d(512, 512, 3, 1, 1), nn.PReLU(),
            nn.MaxPool2d((2, 1), (2, 1)),

            nn.Conv2d(512, 512, (2, 1), 1, 0),
            nn.BatchNorm2d(512), nn.PReLU()
        )

        self.rnn = nn.LSTM(512, nH, bidirectional=True)
        self.linear = nn.Linear(2 * nH, nH)

    def forward(self, input):
        output = self.cnn(input)
        output = output.squeeze(2)
        output = output.permute(2, 0, 1)
        output, _ = self.rnn(output)
        output = self.linear(output)
        return output
````

````python
class Decoder(nn.Module):
    def __init__(self, nIn, nH, nClass):
        super(Decoder, self).__init__()

        self.rnn = nn.LSTM(nIn, nH, bidirectional=True)
        self.linear = nn.Linear(2 * nH, nClass)

    def forward(self, input):
        recurrent, _ = self.rnn(input)
        output = self.linear(recurrent)
        return output
````

````python
nIn    = 512 #размер входного вектора
nH     = 512 #размер скрытого вектора
nClass = len(ABC) #число классов (0 - индекс blank символа)
````

````python
x = torch.randn(20, 1, 512)
y = Decoder(nIn, nH, nClass)(x)

assert y.size() == (20, 1, 23) #(sequence_size, batch_size, hidden_size)
````

#### Общая модель

Теперь объединим `Encoder` и `Decoder` в одну модель `ConvolutionalRecurrentNeuralNetwork`.

**У нас есть `энкодер` и `декодер`. Результатом работы энкодера является карта признаков. На вход декодера требуется подать последовательность векторов. Напишите программный код, который преобразует выход энкодера в вход декодера. Учтите что текст на изображении строго горизонтальный, поэтому нарезаем на последовательность вдоль размерности, отвечающей за ширину карты признаков**

````python
class ConvolutionalRecurrentNeuralNetwork(Module):
    def __init__(self, imgH, nIn, nH, nClass):
        super(ConvolutionalRecurrentNeuralNetwork, self).__init__()
        self.encoder = Encoder(imgH, nH)
        self.decoder = Decoder(nIn, nH, nClass)
        # START YOUR CODE HERE

        # END YOUR CODE HERE
    def forward(self, x):
        # START YOUR CODE HERE
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded
        # END YOUR CODE
````

### <font color='DarkOrange'> **Обработка данных [3 балла]**</font>

Сервисные функции, которые требуется использовать для интерпретации предсказания модели

````python
def pred_to_string(pred, abc):
    """
    Decodes a single prediction into a string using best path decoding.

    Args:
        pred (np.ndarray): 2D array (sequence_length x num_classes) of probabilities.
        abc (str): Alphabet string where each index maps to a character.

    Returns:
        str: Decoded text string.
    """
    seq = []
    for i in range(len(pred)):
        label = np.argmax(pred[i])
        seq.append(label - 1)
    out = []
    for i in range(len(seq)):
        if len(out) == 0:
            if seq[i] != -1:
                out.append(seq[i])
        else:
            if seq[i] != -1 and seq[i] != seq[i - 1]:
                out.append(seq[i])
    out = ''.join([abc[c] for c in out])
    return out
````

````python
def decode(pred, abc):
    """
    Decodes a batch of predictions into a list of strings.

    Args:
        pred (torch.Tensor): Tensor of shape (T, N, C) where:
                             T = time steps,
                             N = batch size,
                             C = number of classes.
        abc (str): Alphabet string.

    Returns:
        List[str]: List of decoded strings, one for each item in the batch.
    """
    pred = pred.permute(1, 0, 2).cpu().data.numpy()
    outputs = []
    for i in range(len(pred)):
        outputs.append(pred_to_string(pred[i], abc[1:]))
    return outputs
````

````python
imgH   = 32
nIn    = 512
nH     = 512
nClass = len(ABC)

crnn      = ConvolutionalRecurrentNeuralNetwork(imgH, nIn, nH, nClass).to(device)
````

````python
crnn.eval()

images, _, _ = next(iter(train_dataloader))
images = images.cuda()
with torch.no_grad():
    pred = crnn(images)
    texts = decode(pred, ABC)

abc_vis = ABC[1:]

fig, ax = plt.subplots(9, 3, figsize=(14, 60))

for i in range(9):
    img = images[i].permute(1, 2, 0).cpu().numpy()
    img = img * np.array([0.229, 0.224, 0.225]) + np.array([0.485, 0.456, 0.406])
    img = np.clip(img, 0, 1)

    ax[i, 0].imshow(img)
    ax[i, 0].axis("off")

    ax[i, 1].text(0.5, 0.5, texts[i], fontsize=18, ha="center", va="center")
    ax[i, 1].axis("off")

    probs = pred[:, i, 1:].softmax(-1).cpu().numpy().T

    ax[i, 2].imshow(probs, aspect="auto")
    ax[i, 2].set_title(texts[i])
    ax[i, 2].set_xlabel("time")
    ax[i, 2].set_ylabel("class")
    ax[i, 2].set_xticks(range(probs.shape[1]))
    ax[i, 2].set_yticks(range(len(abc_vis)))
    ax[i, 2].set_yticklabels(list(abc_vis))

plt.tight_layout()
plt.show()
````
